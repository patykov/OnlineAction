```python
import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
file_path = '/Volumes/UBUNTU 18_0/PAKOVA/outputs/Charades/causal_eval_charades_r50_i3d_nl_32x2_stream_fullyConv_old'

mean_file = file_path + '_mean_results.npy'
median_file = file_path + '_median_results.npy'
max_file = file_path + '_max_pool_results.npy'
gaussian_file = file_path + '_gaussian_results.npy'
```

```python
mean_results = np.load(mean_file, allow_pickle=True)[()]
median_results = np.load(median_file, allow_pickle=True)[()]
max_results = np.load(max_file, allow_pickle=True)[()]
gaussian_results = np.load(gaussian_file, allow_pickle=True)[()]
```

```python
plt.rcParams.update({'font.size': 16})
```

```python
x = np.arange(0, 3.5, 0.5)

fig = plt.figure(figsize=(14, 10))
plt.plot(x, mean_results['map'], marker='o', color='blue', linewidth=2, label="Mean")
plt.plot(x, median_results['map'], marker='o', color='purple', linewidth=2, label="Median")
plt.plot(x, max_results['map'], marker='o', color='red', linewidth=2, label="Max value")
plt.plot(x, gaussian_results['map'], marker='o', color='orange', linewidth=2, label="Gaussian")
plt.legend(loc='lower right', fontsize=20)
plt.xlabel('Delay time (s)', fontsize=20)
plt.ylabel('Per-frame mAP', fontsize=20)
fig.savefig('/Volumes/UBUNTU 18_0/PAKOVA/outputs/Charades/charades_map_delay_funcs.pdf')
plt.show()
```
