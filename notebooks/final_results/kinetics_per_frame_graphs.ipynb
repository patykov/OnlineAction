```python
import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
dir_name = "/Volumes/UBUNTU 18_0/PAKOVA/outputs/Kinetics/"

baseline8_centerCrop_dir = "causal_eval_kinetics_r50_i3d_baseline8_stream_centerCrop"
nl8_centerCrop_dir = "causal_eval_kinetics_r50_i3d_pred_nonlocal_8x8_valCenterCrop"
baseline32_centerCrop_dir = "causal_eval_kinetics_r50_i3d_baseline32_stream_centerCrop"
nl32_centerCrop_dir = "causal_eval_kinetics_r50_i3d_nl32_stream_centerCrop"
```

```python
results_baseline8_centerCrop = np.load(dir_name + baseline8_centerCrop_dir + '_results.npy')
results_nl8_centerCrop = np.load(dir_name + nl8_centerCrop_dir + '_results.npy')
results_baseline32_centerCrop = np.load(dir_name + baseline32_centerCrop_dir + '_results.npy')
results_nl32_centerCrop = np.load(dir_name + nl32_centerCrop_dir + '_results.npy')
```

```python
results_baseline8_centerCrop[:, 0]
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
y = np.arange(10, 110, 10)

fig = plt.figure(figsize=(10, 7))
plt.plot(y, results_baseline8_centerCrop[1:, 0], marker='o', color='blue', linewidth=2, label="baseline_8x8")
plt.plot(y, results_baseline8_centerCrop[1:, 1], marker='o', color='blue', linewidth=2, linestyle='dashed')
plt.plot(y, results_nl8_centerCrop[1:, 0], marker='o', color='purple', linewidth=2, label="nonlocal_8x8")
plt.plot(y, results_nl8_centerCrop[1:, 1], marker='o', color='purple', linewidth=2, linestyle='dashed')
plt.plot(y, results_baseline32_centerCrop[1:, 0], marker='^', color='green', linewidth=2, label="baseline_32x2")
plt.plot(y, results_baseline32_centerCrop[1:, 1], marker='^', color='green', linewidth=2, linestyle='dashed')
plt.plot(y, results_nl32_centerCrop[1:, 0], marker='^', color='orange', linewidth=2, label="nonlocal_32x2")
plt.plot(y, results_nl32_centerCrop[1:, 1], marker='^', color='orange', linewidth=2, linestyle='dashed')
plt.legend(loc='lower right', fontsize=14)
plt.xlabel('Percentage of frames seen', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
fig.savefig('/Volumes/UBUNTU 18_0/PAKOVA/outputs/Kinetics/kinetics_per_frame.pdf')
plt.show()
```
