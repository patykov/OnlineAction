```python
import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
dir_name = "/Volumes/UBUNTU 18_0/PAKOVA/outputs/Kinetics/"

baseline8_centerCrop_dir = "causal_eval_kinetics_r50_i3d_baseline_8x8_centerCrop"
baseline8_fullyConv_dir = "causal_eval_kinetics_r50_i3d_baseline_8x8_fullyConv"
nl8_centerCrop_dir = "causal_eval_kinetics_r50_i3d_nonlocal_8x8_centerCrop"
nl8_fullyConv_dir = "causal_eval_kinetics_r50_i3d_nonlocal_8x8_fullyConv"
```

```python
results_baseline8_centerCrop = np.load(dir_name + baseline8_centerCrop_dir + '_results.npy')
results_baseline8_fullyConv = np.load(dir_name + baseline8_fullyConv_dir + '_results.npy')
results_nl8_centerCrop = np.load(dir_name + nl8_centerCrop_dir + '_results.npy')
results_nl8_fullyConv = np.load(dir_name + nl8_fullyConv_dir + '_results.npy')
```

```python
results_baseline8_centerCrop
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
y = np.arange(10, 110, 10)

fig = plt.figure(figsize=(14, 10))
plt.plot(y, results_baseline8_centerCrop[2:, 0], marker='o', color='blue', linewidth=2, label="baseline8_centerCrop")
plt.plot(y, results_baseline8_centerCrop[2:, 1], marker='o', color='blue', linewidth=2, linestyle='dashed')
plt.plot(y, results_baseline8_fullyConv[2:, 0], marker='o', color='purple', linewidth=2, label="baseline8_fullyConv")
plt.plot(y, results_baseline8_fullyConv[2:, 1], marker='o', color='purple', linewidth=2, linestyle='dashed')
plt.plot(y, results_nl8_centerCrop[2:, 0], marker='^', color='green', linewidth=2, label="nonlocal8_centerCrop")
plt.plot(y, results_nl8_centerCrop[2:, 1], marker='^', color='green', linewidth=2, linestyle='dashed')
plt.plot(y, results_nl8_fullyConv[2:, 0], marker='^', color='orange', linewidth=2, label="nonlocal8_fullyConv")
plt.plot(y, results_nl8_fullyConv[2:, 1], marker='^', color='orange', linewidth=2, linestyle='dashed')
plt.legend(loc='lower right', fontsize=14)
plt.xlabel('Percentage of frames seen', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
fig.savefig('/Volumes/UBUNTU 18_0/PAKOVA/outputs/Kinetics/kinetics_per_frame.pdf')
plt.show()
```

```python
y = np.arange(10, 110, 10)

fig = plt.figure(figsize=(12, 7))
plt.plot(y, results_baseline8_centerCrop[2:, 0], marker='o', color='blue', linewidth=2, label="baseline8_centerCrop", linestyle='dashed')
# plt.plot(y, results_baseline8_centerCrop[2:, 1], marker='o', color='blue', linewidth=2, linestyle='dashed')
plt.plot(y, results_baseline8_fullyConv[2:, 0], marker='o', color='purple', linewidth=2, label="baseline8_fullyConv")
# plt.plot(y, results_baseline8_fullyConv[2:, 1], marker='o', color='purple', linewidth=2)
plt.plot(y, results_nl8_centerCrop[2:, 0], marker='^', color='green', linewidth=2, label="nonlocal8_centerCrop", linestyle='dashed')
# plt.plot(y, results_nl8_centerCrop[2:, 1], marker='^', color='green', linewidth=2, linestyle='dashed')
plt.plot(y, results_nl8_fullyConv[2:, 0], marker='^', color='orange', linewidth=2, label="nonlocal8_fullyConv")
# plt.plot(y, results_nl8_fullyConv[2:, 1], marker='^', color='orange', linewidth=2)
plt.legend(loc='lower right', fontsize=14)
plt.xlabel('Percentage of frames seen', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
fig.savefig('/Volumes/UBUNTU 18_0/PAKOVA/outputs/Kinetics/kinetics_per_frame_top1.pdf')
plt.show()
```
